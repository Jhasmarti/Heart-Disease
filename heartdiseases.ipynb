# Machine Learning Classification

```python
import numpy as np
import pandas as pd
#data
df=pd.read_csv(r"C:\Users\91991\Downloads\archive (1)\heart.csv")
#independent and dependent variable
x=df.drop('target',axis=1)
y=df['target']
def LogisticRegression():
    print("*****************************LogisticRegression********************************************")
    ##Training testing data
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
     ##MAchine learnig alogrithm
    from sklearn.linear_model import LogisticRegression
    model=LogisticRegression()
    model.fit(x_train,y_train)
    accuracy=model.score(x_test,y_test)
    print(accuracy)
    print("*****************************Predication on new data********************************************")
    input_data=(59,1,1,140,221,0,1,164,1,0.0,2,0,2)
    pred=np.array(input_data)
    print("New patient data:",pred)
    predict=pred.reshape(1,-1)
    print("After Reshape data ",predict)
    prediction=model.predict(predict)
    if (prediction[0]==0):
        print("Patient has no heart diseases")
    else:
        print("Heart patient")
    
```

```python
import numpy as np
import pandas as pd
#data
df=pd.read_csv(r"C:\Users\91991\Downloads\archive (1)\heart.csv")
#independent and dependent variable
x=df.drop('target',axis=1)
y=df['target']
def SVC():
    print("*****************************Support Vector Machine********************************************")
    ##Training testing data
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
     ##MAchine learnig alogrithm
    from sklearn.svm import SVC
    model=SVC(kernel='rbf')
    model.fit(x_train,y_train)
    accuracy=model.score(x_test,y_test)
    print(accuracy)
    print("*****************************Predication on new data********************************************")
    input_data=(59,1,1,140,221,0,1,164,1,0.0,2,0,2)
    pred=np.array(input_data)
    print("New patient data:",pred)
    predict=pred.reshape(1,-1)
    print("After Reshape data ",predict)
    prediction=model.predict(predict)
    if (prediction[0]==0):
        print("Patient has no heart diseases")
    else:
        print("Heart patient")
    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#data
df=pd.read_csv(r"C:\Users\91991\Downloads\archive (1)\heart.csv")
#independent and dependent variable
x=df.drop('target',axis=1)
y=df['target']
def DecisionTree():
    print("*****************************DecisionTree********************************************")
    ##Training testing data
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
     ##MAchine learnig alogrithm
    from sklearn.tree import DecisionTreeClassifier
    model=DecisionTreeClassifier()
    model.fit(x_train,y_train)
    accuracy=model.score(x_test,y_test)
    print(accuracy)
    print("*****************************Predication on new data********************************************")
    input_data=(59,1,1,140,221,0,1,164,1,0.0,2,0,2)
    pred=np.array(input_data)
    print("New patient data:",pred)
    predict=pred.reshape(1,-1)
    print("After Reshape data ",predict)
    prediction=model.predict(predict)
    if (prediction[0]==0):
        print("Patient has no heart diseases")
    else:
        print("Heart patient")
    
```
